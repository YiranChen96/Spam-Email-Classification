```python
#Load data file from local 
import pandas as pd
inputfile ='/Users/chenyang/Desktop/spambase.data.txt'
raw=[]
with open(inputfile,'r') as f:
	for line in f:
		raw.append(list(line.strip('\n').split(',')))
data=pd.DataFrame(raw)
```

```python
# extract x_train and label
x=data.iloc[:,0:57]
y=data.iloc[:,57]
```

```python
#To find the best parameter c,g for this SVM model
from sklearn.model_selection import KFold
from sklearn import svm
from sklearn.metrics import accuracy_score
import numpy as np
X=np.array(x)
Y=np.array(y)
c_range=np.logspace(-2,4,7)
g_range=np.logspace(-4,1,6)
maxscore=0
for c in c_range:
    for g in g_range:
        kf = KFold(n_splits=10,shuffle=True)
        kf.get_n_splits(X)
        scores=[]
        for train, test in kf.split(X):
            x_train,x_val = X[train], X[test]
            y_train,y_val = Y[train],Y[test]
            clf = svm.SVC(gamma=g,kernel='rbf',decision_function_shape='ovr',C=c)
            clf.fit(x_train,y_train)
            y_pred=clf.predict(x_val)
            score=accuracy_score(y_val, y_pred)
            scores.append(score)
        mscore=sum(scores)/len(scores)
        print("c=",c,"g=",g,"score=",mscore)
        if mscore>maxscore:
            maxscore=mscore
            bestc=c
            bestg=g
print("The best C is",bestc,"The best g is",bestg,"achieve an accuracy of",maxscore) 
```

```python
#So we choose g=0.0001 and c=1000 as the SVM parameter and apply the model to each fold
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import confusion_matrix
#Perform 10-fold cross-validation,divide data into 10 sets, take each one as test data and the other 9as training data.
table=[]
kf = KFold(n_splits=10,shuffle=True)
kf.get_n_splits(X)
scores=[]
FNR=[]
FPR=[]
for train, test in kf.split(X):
    x_train,x_val = X[train], X[test]
    y_train,y_val = y[train],y[test]
    clf = svm.SVC(gamma=bestg,kernel='rbf',decision_function_shape='ovr',C=bestc)
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_val)
    score=accuracy_score(y_val, y_pred)
    scores.append(score)
    ConfusionMatrix=confusion_matrix(y_val, y_pred)
    #print(pd.DataFrame(ConfusionMatrix))
    tn, fp, fn, tp =ConfusionMatrix.ravel()
    FalseNegativeRate=float(fn)/(fn+tp)
    FNR.append(FalseNegativeRate)
    FalsePositiveRate=float(fp)/(fp+tn)
    FPR.append(FalsePositiveRate)
    ErrorRate=1-score
    #print(errorRate)
    #print(FalseNegativeRate,FalsePositiveRate)
    table.append([FalsePositiveRate,FalseNegativeRate,ErrorRate])
OverallFNR=sum(FNR)/len(FNR)
OverallFPR=sum(FPR)/len(FPR)
OverallErrorRate=1-sum(scores)/len(scores)#Take the average of ten times' scores as the final result
table.append([OverallFPR,OverallFNR,OverallErrorRate])

    
```

```python
#create table show the result for each fold and the overall accuracy
Table=pd.DataFrame(table)
Table.columns=['FalsePositiveRate','FalseNegativeRate','ErrorRate']
Table.rename({0:'Fold_1',1:'Fold_2',2:'Fold_3',3:'Fold_4',4:'Fold_5',5:'Fold_6',6:'Fold_7',7:'Fold_8',8:'Fold_9',9:'Fold_10',10: 'Overall'}, axis='index')
```
